```python
import pandas as pd
import numpy as np
```

```python
med_data = pd.read_csv('Medicaid_Spending_by_Drug_2021.csv')
```

```python
med_data.head(5)
```

```python
# check if any duplicates 
duplicated = med_data[med_data.duplicated()]
print(duplicated)
```

```python
nan_count = med_data.isna().sum()
print(nan_count)
```

```python
# drop unwanted columns 
med_data = med_data.drop(['Chg_Avg_Spnd_Per_Dsg_Unt_20_21', 'CAGR_Avg_Spnd_Per_Dsg_Unt_17_21'], axis=1)
```

```python
# remove dollar signs & commas
med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']] = med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']].apply(lambda x: x.str.replace('$',''))
med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']] = med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']].apply(lambda x: x.str.replace(',',''))


# fill NAs with Zeros
med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']] = med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']].fillna(0)


# change to float 
med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']] = med_data[['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']].astype(float)
med_data.head(5)
```

```python
# calculate yearly moving total spending 
cols = ['Tot_Spndng_2017', 'Tot_Spndng_2018', 'Tot_Spndng_2019', 'Tot_Spndng_2020', 'Tot_Spndng_2021']
totalsp_rolling = med_data[cols].rolling(window=2)
ma_med_data_total = totalsp_rolling.mean()
print(ma_med_data_total)
```

```python
#write cleaned data to sheet
med_data.to_csv('meddata.csv')
ma_med_data_total.to_csv('runningdata.csv')
```
